```python
import pandas as pd

# 1. 엑셀 파일 불러오기
file_path = "/content/지역별 인구 및 인구밀도_20250828232703.xlsx"
df = pd.read_excel(file_path, sheet_name='지역별 인구 및 인구밀도(e-나라지표)', skiprows=28)
df
```

```python
# 2. 컬럼 정리
df = df.rename(columns={"Unnamed: 0": "지역"})
# 불필요한 NaN 제거
df = df.dropna(subset=["지역"])
df
```

```python
# 3. 불필요한 행 제거 (출처, 단위, 주석 제거)
df = df.loc[:19, :].copy()
df
```

```python
# 3. 숫자형 변환 (천 단위 구분 기호 제거)
for col in df.columns[1:]:
    df[col] = df[col].astype(str).str.replace(",", "").astype(float)
df
```

```python
# 4. 연도별 총인구 및 평균 인구 변화 추이
years = [2020, 2021, 2022, 2023, 2024]
total_pop = {year: df[str(year)].sum() for year in years}
total_pop
```

```python
# 5. 지역별 인구 증감률 (2020→2024)
df["증감률(2020~2024)"] = ((df["2024"] - df["2020"]) / df["2020"]) * 100
df
```

```python
# 6. 인구 감소 지역 TOP 5
top5_decline = df.sort_values("증감률(2020~2024)").head(5)
top5_decline
```

```python
print("연도별 총 인구:", total_pop)
print("\n인구 감소 지역 TOP 5:")
print(top5_decline[["지역", "증감률(2020~2024)"]])
```
